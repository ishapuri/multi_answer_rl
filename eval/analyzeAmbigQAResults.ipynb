```python
from datasets import load_dataset

# Load the dataset from HuggingFace Hub
dataset = load_dataset("mehuldamani/qwen3_8b_ambigQA_rlcr_single_passk")

# Optionally print out the keys (splits) and the first example for inspection
print("Dataset splits:", dataset.keys())
print("First example in 'train' split:", dataset["test"][0])
```

```python
# Print all the column names in the 'test' split
print("Column names:", dataset["test"].column_names)
```

```python
# Assuming "RLCR Single temp 0.7-evals" is a column in the dataset,
# and each entry is a list of numbers

# Extract the relevant column from the "test" split
eval_lists = dataset["test"]["RLCR Single temp 0.7-evals"]

# For each list, compute its sum
row_sums = [sum(row) for row in eval_lists]

# Compute the average of these sums
average_sum = sum(row_sums) / len(row_sums) if row_sums else 0

print("Average of the sum of each list in 'RLCR Single temp 0.7-evals':", average_sum)
```

```python
# Assuming "RLCR Single temp 0.7-evals" is a column in the dataset,
# and each entry is a list of numbers

# Extract the relevant column from the "test" split
eval_lists = dataset["test"]["RLCR Multiple temp 0.7-candidate_correctness"]

# For each list, compute its sum
row_sums = [sum(row) for row in eval_lists]

# Compute the average of these sums
average_sum = sum(row_sums) / len(row_sums) if row_sums else 0

print("Average of the sum of each list in 'RLCR Multiple temp 0.7-candidate_correctness':", average_sum)
```

```python
# Efficiently compare the average number of correct answers for Single vs Multi runs

# Specify the columns to compare and their friendly names
columns_to_compare = [
    ("RLCR Single temp 0.7-evals", "Single"),
    ("RLCR Single temp 0.0-evals", "Single-temp0"),
    ("RLCR Multiple temp 0.7-candidate_correctness", "Multi"),
    ("RLCR Multiple temp 0.0-candidate_correctness", "Multi-temp0")
]

results_sum = {}
for col, label in columns_to_compare:
    # Each entry in the column is a list of numbers
    eval_lists = dataset["test"][col]
    # For each list, compute its sum (number of correct answers for that example)
    row_sums = [sum(row) for row in eval_lists if len(row) > 0 and len(row)==3]
    # Compute the average number of correct answers
    average_sum = sum(row_sums) / len(row_sums) if row_sums else 0
    results_sum[label] = average_sum

# Print out the comparison
for label in results_sum:
    print(f"Average number of correct answers in '{label}' run:", results_sum[label])
```

```python
single_which_answers = qwen3_8b_ambigQA_single_dataset['test']['RLCR Single temp 0.7-which_answers']
multi_which_answers = qwen3_8b_ambigQA_multi_dataset['test']['RLCR Multiple temp 0.7-which_answers']

print(single_which_answers[0])
print(multi_which_answers[0])
```

```python
def find_average_number_of_unique_correct_answers_per_question(which_answers):
    which_answers_name = "single" if which_answers == single_which_answers else "multi"

    num_unique_correct_answers = 0
    total_num_questions = 0

    for row in which_answers:
        total_num_questions += 1
        unique_correct_idxs = []
        for val in row:
            if val >= 0:
                if val not in unique_correct_idxs:
                    unique_correct_idxs.append(val)
                    num_unique_correct_answers += len(unique_correct_idxs)

    avg_unique_correct_answers = num_unique_correct_answers / total_num_questions
    print(f"Average number of unique correct answers per question for {which_answers_name}: {avg_unique_correct_answers:.3f}")

find_average_number_of_unique_correct_answers_per_question(single_which_answers)
find_average_number_of_unique_correct_answers_per_question(multi_which_answers)
```

```python
def find_average_number_of_unique_correct_answers_per_correct_question(which_answers):
    which_answers_name = "single" if which_answers == single_which_answers else "multi"
    num_unique_correct_answers = 0
    total_num_questions = 0

    for row in which_answers:
        if len([v for v in row if v >= 0]) < 1:
            continue
        total_num_questions += 1
        unique_correct_idxs = []
        for val in row:
            if val >= 0:
                if val not in unique_correct_idxs:
                    unique_correct_idxs.append(val)
                    num_unique_correct_answers += len(unique_correct_idxs)

    avg_unique_correct_answers = num_unique_correct_answers / total_num_questions
    print(f"Average number of unique correct answers per correct question for {which_answers_name}: {avg_unique_correct_answers:.3f}")

find_average_number_of_unique_correct_answers_per_correct_question(single_which_answers)
find_average_number_of_unique_correct_answers_per_correct_question(multi_which_answers)
```
